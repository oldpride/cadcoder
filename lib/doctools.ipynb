```python
import re

# reduce floating point numbers to 6 decimal places
test_lines = [
    '9.123456789',
    '( -9.123456789, 10.123456789, 0.000123456789)'
]

for line in test_lines:
    line2 = re.sub(r'\d+\.\d+', lambda x: str(round(float(x.group()), 6)), line)
    print(f"line={line:30s} line2={line2:30s}")
```

```python
import re

test_lines = [
    '0.000',
    '-0.000',
    '10.00',
    '(-0.000, 0.000)'
]

for line in test_lines:
    line2 = re.sub(r'\b0.0+\b', '0', line)
    line3 = re.sub(r'-0\b', '0', line2)
    print(f"line={line:6s} line2={line2:6s} line3={line3:6s}")
```

```python
# 6.275484e-16, 1e-6, ... should be considered as 0
test_lines = [
    '6.275484e-16',
    '1e-6',
    '-1e-6',
    '(6.275484e-16, 1.4e-4, 1e-6, -1e-4)'
    
]
for line in test_lines:
    line2 = re.sub(r'[-+]?\d*(\.\d+)?e[-]?\d+', lambda x: '0' if abs(float(x.group())) < 1e-4 else x.group(), line)
    print(f"line={line}, line2={line2}")
```
